*****

Calculating per day percentage return.

```python
ind_return=(ind_data/ind_data.shift(1))-1
ind_return.tail()
```

#Estimated the average annual return of each index assuming their were 250 working days in market.

```python
annual_returns=ind_return.mean()*250*100
print(annual_returns)
```
